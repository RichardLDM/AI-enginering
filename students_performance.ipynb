```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Student_performance_data _.csv")

print("Dataset shape (rows, columns):", df.shape)
print("Column name:", df.columns.tolist())

print(df.head())

print("Summary statistics:")
print(df.describe())

print("Data types and null info:")
print(df.info())
```

```python
print("Average score by columns:")
df[["GPA", "Volunteering"]].mean()
```

```python
print("MAX score by columns:")
df[["GPA", "Volunteering"]].max()
```

```python
print("MAX score by columns:")
df[["GPA", "Volunteering"]].min()
```

```python
# Correlation
print("Correlation")
df[["GPA", "Volunteering"]].corr()
```

```python
df["Total score"] = df[["GPA", "GradeClass"]].sum(axis=1)
df["average score"] = df["Total score"] / 2
df["passed"] = np.where(df["average score"] >= 2, "Pass", "Fail")

print("New columns preview(total, avergae, passed):")
print(df[["Total score", "average score", "passed"]])
```

```python
test_prep_group = df.groupby("Tutoring") [["GPA", "GradeClass"]].mean()
print("Grouped by Tutoring")
print(test_prep_group)

# Gender

gender_avg = df.groupby("Gender") [["GPA", "GradeClass"]].mean()
print("Grouped by gender")
print(gender_avg)

# Who passed vs failed
pass_rate = df["passed"].value_counts(normalize=True) * 100
print("Passed vs failed")
print(pass_rate)
```

```python
# Convert scores to Numpy array
scores = df[["GPA", "GradeClass"]].values

# Mean, Std Deviation
print("Numpy Mean:", np.mean(scores, axis=0))
print("Numpy Std Deviation:", np.std(scores, axis=0))

# Z-score (normalize)
z_scores = (scores - np.mean(scores, axis=0)) / np.std(scores, axis=0)
print("Z-score:")
print(z_scores[:5])
```

```python
# How many students scored above 3 avaerge?
top_scorers = df[df["average score"] >= 2.5]
print("Number of students with an average score of 3 or higher:", len(top_scorers))

# How many failed
failed_students = df[df["passed"] == "Fail"]
print("Number of students who failed:", len(failed_students))
```
